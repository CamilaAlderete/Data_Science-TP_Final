### Gráficos

```python
#Importar librerias graficas de python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
#Gráfico de Barras del número de Aplicaciones de cada Categoría
plt.figure(figsize = (65, 30))
grafico = sns.countplot(x = 'Category', data = dataset)
grafico.set_title("Número de Aplicaciones de cada Categoría", fontsize = 70)
plt.xlabel("Category" , fontsize = 50)
plt.ylabel("Count" , fontsize = 50)
```

```python
#Gráfico tipo Pie del número de Aplicaciones de cada Categoría, muestra un gráfico más legible que el anterior

import numpy as np
# eliminamos las categorias con menos porcentaje para obtener un gráfico sin solapamientos, dejamos solo las categorías más importantes

eliminar = ['Weather',                
'Dating',                       
'Casino',                       
'Parenting',                   
"Libraries & Demo",             
'Music',                        
'Comics',
'Beauty',
'Trivia',
"Role Playing",
'Racing',
'Board',
'Word',
'Card',
'Strategy',
"Video Players & Editors" ]

# contamos la cantidad de aplicaciones existentes de cada categoría
data = dataset.Category.value_counts().drop(index = eliminar).array

# obtenemos los nombres de las categorías
categories = dataset.Category.value_counts().drop(index = eliminar).index.array

fig, ax = plt.subplots(figsize=(15, 10))
colors = sns.color_palette('cool_r')

wedges, texts  = ax.pie(data, startangle=0, wedgeprops={'edgecolor': 'white'}, colors = colors)
  
# para mostrar el porcentaje en el gráfico
perc = [str(round(e / s * 100., 1)) + '%' for s in (sum(data),) for e in data]

kw = dict(arrowprops=dict(arrowstyle="-"), zorder=0, va="center")

# para colocar el nombre de cada categoría con su porcentaje de una manera legible
for i, p in enumerate(wedges):
    ang = (p.theta2 - p.theta1) / 2. + p.theta1
    y = np.sin(np.deg2rad(ang))
    x = np.cos(np.deg2rad(ang))
    yc = np.arcsin(y) / (np.pi / 2)
    horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
    connectionstyle = f'angle,angleA=0,angleB={ang}'
    kw["arrowprops"].update({"connectionstyle": connectionstyle})
    ax.annotate(categories[i] + ' ' + perc[i], xy = (0.8 * x, 0.8 * y), xytext = ((1.0 + (i % 2) * 0.4) * np.sign(x), 1.4 * yc),
                horizontalalignment = horizontalalignment, fontsize ='x-large', **kw)
    
plt.title("Gráfica del Porcentaje de Aplicaciones de cada Categoría", fontsize=15)

plt.show()
```

```python
#Gráfico tipo Mapa de Calor de las correlaciones entre variables del dataset
plt.figure(figsize = (12, 8))
grafico = sns.heatmap(dataset.corr(), annot = True, cmap="Blues")
grafico.set_title("Mapa de Calor de las Correlaciones entre Características de las Aplicaciones\n", fontsize = 15)
```

#####  <blockquote>   ¡¡Posible redundacia de datos!! 
#### 2.  Notamos que la variable "Rating Count" esta Medianamente Correlacionado con las variables  "Minimum Installs" y "Maximum Installs"

```python
#Gráfico de Conteo para mostrar el número de aplicaciones con cada tamanho

#Dividir el dataframe para poder realizar un grafico legible, un grafico de un sector del dataframe, ya que existe un rango...
            # ... de datos muy grande en la columna Size

#Creamos 2 dataframes, el size_1 de 2OO filas y el size_2 de 400 filas para Analizar estos datos 
size_1= dataset.iloc[:200,:]
size_2 = dataset.iloc[200:800:,:]

#PARTE 1:

#Dibujar el sector 1 de la division
plt.figure(figsize = (22, 10))
ax = sns.countplot(x = 'Size', data = size_1)

#Rotar las etiquetas del eje x
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="right")
plt.tight_layout()

#Titulo y nombre de los ejes
plt.xlabel("\n\nSize" , fontsize = 25)
plt.ylabel("Count" , fontsize = 25)
plt.title("Gráfica de Conteo de Aplicaciones Clasificadas por 'Size'", fontsize = 25)

plt.show()
```

```python
#PARTE 2:

#Dibujar el sector 2 de la division
plt.figure(figsize = (30, 20))
ax = sns.countplot(x = 'Size', data = size_2)

#Rotar las etiquetas del eje x
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="right")
plt.tight_layout()

#Titulo y nombre de los ejes
plt.xlabel("\nSize" , fontsize = 25)
plt.ylabel("Count" , fontsize = 25)
plt.title("Gráfica de Conteo de Aplicaciones Clasificadas por 'Size'", fontsize = 25)

plt.show()
```
